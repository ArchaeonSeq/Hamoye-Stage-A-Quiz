# QUESTION 14

```python
ds_corr= ds.corr()
ds_corr
```

```python
listd =['Element Code', 'Area Code', 'Item Code']
```

```python
ds_corr_a = ds_corr.loc[~ds_corr.index.isin (listd)]['Element Code'].sort_values(ascending =False).head(1)
print("The year with the highest correlation with Element Code is shown below:" , ds_corr_a.index)
ds_corr_a
```

```python
#Or we could use this method.
ds_corr_b = ds_corr.drop(listd)
ds_corr_b= ds_corr_b["Element Code"].sort_values(ascending=False).head(1)
print("The year with the highest correlation with Element Code is shown below:" , ds_corr_b.index)
ds_corr_b
```
